```python
import pandas as pd

catboost = pd.read_csv('catboost_results.csv')
patchtst = pd.read_csv('patchtst_results.csv')
```

```python
catboost_summary = catboost.groupby('method')['sMAPE'].agg(['mean', 'std']).round(2)
patchtst_summary = patchtst.groupby('method')['sMAPE'].agg(['mean', 'std']).round(2)

print(catboost_summary)
print(patchtst_summary)
```

```python
results = pd.DataFrame({
    'None': [
        catboost[catboost['method']=='none']['sMAPE'].mean(),
        patchtst[patchtst['method']=='none']['sMAPE'].mean()
    ],
    'Standard': [
        catboost[catboost['method']=='standard']['sMAPE'].mean(),
        patchtst[patchtst['method']=='standard']['sMAPE'].mean()
    ],
    'Robust': [
        catboost[catboost['method']=='robust']['sMAPE'].mean(),
        patchtst[patchtst['method']=='robust']['sMAPE'].mean()
    ]
}, index=['CatBoost', 'PatchTST'])

results
```
